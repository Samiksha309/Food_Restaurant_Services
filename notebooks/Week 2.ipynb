```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("processed_sales.csv")

df["Date"] = pd.to_datetime(df["date"])

df.set_index("Date", inplace=True)

df.head()
```

```python
df["year"] = df.index.year

df["month"] = df.index.month

df["quarter"] = df.index.quarter

df["week"] = df.index.isocalendar().week.astype(int)

df["day"] = df.index.day

df["dayOfWeek"] = df.index.dayofweek

df["dayName"] = df.index.day_name()
```

```python
df["IsWeekend"] = np.where(df["dayOfWeek"]>=5,1,0)
```

```python
import holidays

india = holidays.India()

df["IsHoliday"] = df.index.map(
    lambda x: 1 if x in india else 0
)
```

```python
df["Lag_1"] = df["quantity_sold"].shift(1)

df["Lag_7"] = df["quantity_sold"].shift(7)

df["Lag_14"] = df["quantity_sold"].shift(14)

df["Lag_30"] = df["quantity_sold"].shift(30)
```

```python
df["RollingMean7"] = df["quantity_sold"].rolling(7).mean()

df["RollingMean14"] = df["quantity_sold"].rolling(14).mean()

df["RollingMean30"] = df["quantity_sold"].rolling(30).mean()

df["RollingStd7"] = df["quantity_sold"].rolling(7).std()
```

```python
df["EMA7"] = df["quantity_sold"].ewm(span=7).mean()

df["EMA14"] = df["quantity_sold"].ewm(span=14).mean()
```

```python
df["ExpandingMean"] = df["quantity_sold"].expanding().mean()
```

```python
df = df.dropna()

df.head()
```

```python
train_size = int(len(df)*0.8)

train = df.iloc[:train_size]

test = df.iloc[train_size:]

print(train.shape)

print(test.shape)
```
